```python
from sklearn import datasets
```

```python
from matplotlib import pyplot as plt
```

```python
digits = datasets.load_digits()
# Take the first 500 data points: it's hard to see 1500 points
X = digits.data[:500]
y = digits.target[:500]
```

```python
from sklearn.manifold import TSNE
```

```python
target_ids = range(len(digits.target_names))

for iterations in [250, 1000, 2000]:
    tsne = TSNE(n_components=2, random_state=0,  n_iter = iterations, perplexity = 30)
    X_2d = tsne.fit_transform(X)
    plt.figure(figsize=(6, 5))
    colors = 'r', 'g', 'b', 'c', 'm', 'y', 'k', 'w', 'orange', 'purple'
    for i, c, label in zip(target_ids, colors, digits.target_names):
        plt.scatter(X_2d[y == i, 0], X_2d[y == i, 1], c=c, label=label)
    print("Scatter plot for no. of iterations =",iterations)
    plt.legend()
    plt.show()
```

# (a) Fix perplexity to 30, repeat the algorithm with a number of iterations = 250, 1000 and 2000. 
- We observe that the scatter plots, experiments with no. of iterations = 1000 and 2000 produce the same results because t-SNE captures structure in the sense that neighboring points in the input space will tend to be neighbors in the low dimensional space. 
- Therefore, same results are produced. Also choosing hyperparameter depends on given dataset.

(b) This is because to perform t-SNE repeatedly with different values of hyperparameters before arriving at the suitable graph for visualization and produces different results.
